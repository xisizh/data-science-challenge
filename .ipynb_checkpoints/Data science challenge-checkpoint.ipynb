# Tagup data science challenge

ExampleCo, Inc has a problem: maintenance on their widgets is expensive. They have contracted with Tagup to help them implement predictive maintenance. They want us to provide a model that will allow them to prioritize maintenance for those units most likely to fail, and in particular to gain some warning---even just a few hours!---before a unit does fail.

They collect two kinds of data for each unit. First, they have a remote monitoring system for the motors in each unit, which collects information about the motor (rotation speed, voltage, current) as well as two temperature probes (one on the motor and one at the inlet). Unfortunately, this system is antiquated and prone to communication errors, which manifest as nonsense measurements. Second, they have a rule-based alarming system, which can emit either warnings or errors; the system is known to be noisy, but it's the best they've got. 

They have given us just over 100MB of historical remote monitoring data from twenty of their units that failed in the field. The shortest-lived units failed after a few days; the longest-lived units failed after several years. Typical lifetimes are on the order of a year. This data is available in .csv files under `data/train` in this repository. In addition, they have provided us with operating data from their thirty active units for the past month; this data is available under `data/test` in this repository.

```python
# To help you get started...
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

from datetime import datetime

from xgboost import XGBClassifier

import pywt
```

```python
# help functions

def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))


# help function for removing noise
def denoise_wavelet(time_series, sigma):
    #noiseSigma = np.median(np.abs(time_series))/0.6745
    noiseSigma = sigma
    wavelet = pywt.Wavelet('haar')
    WaveletCoeffs = pywt.wavedec(time_series, wavelet)
    uthresh = noiseSigma*np.sqrt(2*np.log2(len(time_series)))
    WaveletCoeffs[1:] = ( pywt.threshold(i, value=uthresh, mode="soft" ) for i in WaveletCoeffs[1:] )
    sig_new = pywt.waverec(WaveletCoeffs, wavelet)
    
    return sig_new


def clean_features(feature, threshold):
    # replace abnormal large value by median
    # return: ndarray
    m = 10**6
    feature[feature<-threshold] = -m
    feature[feature>threshold] = m
    
    return feature


def cleaned_data(data):
    # return: dataframe
    rms_clean = pd.DataFrame()
    rms_clean['rpm'] = denoise_wavelet(clean_features(data.rpm, 2000), 8)[:-1]
    rms_clean['motor_voltage'] = denoise_wavelet(clean_features(data.motor_voltage, 400), 5)[:-1]
    rms_clean['motor_current'] = denoise_wavelet(clean_features(data.motor_current, 60), 1)[:-1]
    rms_clean['motor_temp'] = denoise_wavelet(clean_features(data.motor_temp, 250), 1)[:-1]
    rms_clean['inlet_temp'] = denoise_wavelet(clean_features(data.inlet_temp, 180), 1)[:-1]
    rms_clean['num_warnings'] = data.num_warnings
    
    return rms_clean


def train_data_processing(length1=1000, length2=300):
    # use measurements in time length1 to predict if unit fail in the next time length2
    # length2 should be larger than 0
    # return: precessed data list[], [ndarray, ndarray]
    
    precessed_data = []
    y = []
    
    for i in range(20):
        if i < 10:
            rms_filename = 'data/train/unit000'+str(i)+'_rms.csv'
            alarms_filename = 'data/train/unit000'+str(i)+'_alarms.csv'
        else:
            rms_filename = 'data/train/unit00'+str(i)+'_rms.csv'
            alarms_filename = 'data/train/unit00'+str(i)+'_alarms.csv'
            
        alarms = pd.read_csv(alarms_filename, header=None, names=["timestamp", "message"])
        rms = pd.read_csv(rms_filename).fillna(0)
        # add feature 'num_warnings' to data
        rms['num_warnings'] = num_warnings(rms, alarms)
        rms = rms.drop(['timestamp'], axis=1)
        # remove noise 
        rms = cleaned_data(rms).values
        
        for k in range(length2):
            precessed_data.append(rms[len(rms)-length1-k:len(rms)-k].reshape(-1))
            y.append(1)
        for k in range(length2, length2*2):
            precessed_data.append(rms[len(rms)-length1-k:len(rms)-k].reshape(-1))
            y.append(0)
            
    return [np.array(precessed_data), np.array(y)]


def test_data_processing(length1=1000, length2=10):
    # return precessed data, [ndarray, ndarray] or ndarray
    
    precessed_data = []
    if length2 != 0: y = []
    for i in range(20,50):
        rms_filename = 'data/test/unit00'+str(i)+'_rms.csv'
        alarms_filename = 'data/test/unit00'+str(i)+'_alarms.csv'
        
        rms = pd.read_csv(rms_filename)
        alarms = pd.read_csv(alarms_filename, header=None, names=["timestamp", "message"])
        rms['num_warnings'] = num_warnings(rms, alarms)
        rms = rms.drop(['timestamp'], axis=1).fillna(0)
        # remove noise 
        rms = cleaned_data(rms).values
        
        if length2 != 0:
            for k in range(length2):
                precessed_data.append(rms[k:length1+k].reshape(-1))
                y.append(0)
        else:
            precessed_data.append(rms[len(rms)-length1:].reshape(-1))
    
    return [np.array(precessed_data), np.array(y)] if length2!=0 else np.array(precessed_data)
```

### Load and process data

```python
def load_rms(filename):
    return pd.read_csv(filename, index_col="timestamp")
def load_alarms(filename):
    return pd.read_csv(filename, header=None, names=["timestamp", "message"], index_col="timestamp")

rms = load_rms('data/train/unit0000_rms.csv')
alarms = load_alarms('data/train/unit0013_alarms.csv')
rms.loc["2005-08-01":"2005-09-01"].plot(ylim=(-10, 1500))
rms.loc["2005-08-01":"2005-08-02"].plot(ylim=(-10, 1500))

display(rms.describe())
display(alarms.describe())
```

There are clearly some extreme values that are nonsense measurements. After removing noise, 'rpm' became smoother.

```python
temp = rms.rpm
# replace large value
value = 10**6
temp[temp<-2000] = -value
temp[temp>2000] = value

rms['rpm_clean'] = denoise_wavelet(temp,8)[:-1]

rms.loc["2005-08-01":"2005-09-01"].plot(ylim=(-10, 1500))
rms.loc["2005-08-01":"2005-08-02"].plot(ylim=(-10, 1500))
```

```python
# both time length are measured in terms of number of timestamps
time_length1 = 1000
time_length2 = 10

start_time = timer(None) # timing starts from this point for "start_time" variable
data, y = train_data_processing(time_length1, time_length2)
timer(start_time) # timing ends here for "start_time" variable

print(data.shape, y.shape)

X = data.copy()
```

### Train classification models

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=42, stratify=y)
print(X_train.shape, y_train.shape)
```

```python
# train random forest using 5-fold corss-validation
randomforest = RandomForestClassifier(n_estimators=100, random_state=42, max_features='sqrt')
randomforest.fit(X_train,y_train)

forest_scores = cross_val_score(randomforest, X_train, y_train, scoring='neg_mean_squared_error', cv=5) 

rf_mse_train = np.mean(-forest_scores)
print('\n The misclassification error of training set is :', rf_mse_train)

rf_pred_val = randomforest.predict(X_val)

rf_mse_val = mean_squared_error(y_val, rf_pred_val)
print('\n The misclassification error of validation set is :', rf_mse_val)


# create confusion matrix
print(classification_report(y_val, rf_pred_val))

print('Confusion Matrix for test data:')
print(confusion_matrix(y_val, rf_pred_val))
```

```python
xgb_cla = XGBClassifier(n_estimators=100, random_state=42, max_depth=3, learning_rate=0.01)
# Add silent=True to avoid printing out updates with each cycle
xgb_cla.fit(X_train, y_train)

xgb_pred_train = xgb_cla.predict(X_train)
xgb_mse_train = mean_squared_error(y_train, xgb_pred_train)
print('\n The misclassification error of training set is :', xgb_mse_train)

xgb_pred_val = xgb_cla.predict(X_val)

xgb_mse_val = mean_squared_error(y_val, xgb_pred_val)
print('\n The misclassification error of test set is :', xgb_mse_val)

# create confusion matrix
print(classification_report(y_val, xgb_pred_val))

print('\n Confusion Matrix for test data:')
print(confusion_matrix(y_val, xgb_pred_val))
```

### Predicting if fail for units in test data

```python
time_length2 = 0
start_time = timer(None) # timing starts from this point for "start_time" variable
X_test_pred = test_data_processing(time_length1, time_length2)
timer(start_time) # timing ends here for "start_time" variable
print(X_test_pred.shape)
```

```python
xgb_pred_test = xgb_cla.predict(X_test_pred)
xgb_pred_test_proba = xgb_cla.predict_proba(X_test_pred)
```

```python
rf_pred_test = randomforest.predict(X_test_pred)
rf_pred_test_proba = randomforest.predict_proba(X_test_pred)
```

```python
idx = []
for i, pred in enumerate(zip(rf_pred_test, xgb_pred_test)):
    if pred[0] and pred[1]: idx.append(i)
probs = []
for i in idx:
    probs.append(np.round((rf_pred_test_proba[i][1]+xgb_pred_test_proba[i][1])/2, 4))
    
print("\n The units are most likely fail with probability of fail \n")
for unit, prob in zip(idx, probs):
    print(unit+20, prob)
```

## Summary

In **data preprocessing** part, first, I loaded all rms and alarms data from `data/train` followed by combining rms and alarms dataframe to make use of information of warnings. A new feature "num_warnings" which is the number of warnings between current timestamp and the previous one timestamp, was created and added to "rms".

Second, I tried to remove the noise for "rpm" and "motor_voltage" to capture the main pattern. I replaced extremely large and small values instead of removing them, because abnormal values indicate abnormal unit status which might cause unit failure. I removed noise for "rpm" and "motor_voltage". In the plot below, the curve of 'cleaned' "rpm" became smoother. 

I decided to use classification models to predict unit failures within given a time window. Even though these methods can't provide very accurate lifetime predictions for units far in the future, they can let us know if the units will fail in a short time period so as to help ExampleCo with their maintenance. So, third, I constructed a balanced training set and test set. Response variable y represents if the unit fail after a short time period, and features are reconstructed subsets, which are historical measurements in a time window, from "rms" dataframe. The length of time period is measured in terms of number of timestamps.

In model fitting part, after splitting processed data into training and validating set, I trained randomforest and xgboost classifiers on training data and evaluated the model performance on validating set. The goal is using information in a time period (1000 timestamps) to predict if unit fail 'soon' (in the next 10 timestamps). As, a result, both classifiers performed well with **low error (0.09 for randomforest and 0.16 for xgboost)** and **high recall (0.87 for both classifiers)** and precision on validating set! In this case, we want high recall, which is among all units are failed, the proportion classifier correctly predicted fail.

In **prediction** part, after loading all data from `data/test` and processing test data in the same way, I used a subset of test data (the most recent 1000 timestamps) to predict if unit fail in a short time period (next 10 timestamps). The two classifiers gave different predictions for the 30 units. In order to find the units are most likely to fail, I chose the units which predicted to fail in the short future by both classifiers.

Since I used balanced data by taking a small subset from original data set for classification task, I lost many information of data. If I had more time, I would make use of more information and try different models. Not just focus on predict fail or not in short future.
